```python
import polars as pl
from great_tables import GT
from great_tables.data import exibble
```

## data_color

```python
df = pl.from_pandas(exibble)
df
```

```python
# default
(GT(df).data_color())
```

```python
# palette
(GT(df).data_color(columns="row", palette="Blues"))
```

```python
# palette - multiple colors
(
    GT(df).data_color(
        columns="row", palette=["rebeccapurple", "white", "orange"]
    )
)
```

```python
# domain
(GT(df).data_color(columns="num", palette="Oranges", domain=[0, 5600]))
```

```python
# na_color
(
    GT(df).data_color(
        columns="num",
        palette="Oranges",
        domain=[0, 5600],
        na_color="lightgreen",
    )
)
```

```python
# multiple `GT.data_color()`
(
    GT(df)
    .data_color(
        columns="num",
        palette="Oranges",
        domain=[0, 5600],
        na_color="lightgreen",
    )
    .data_color(
        columns="currency",
        palette="Purples",
        domain=[0, 50],
        na_color="lightgreen",
    )
)
```

```python
# rows
(
    GT(df).data_color(
        columns="currency",
        rows=pl.col("currency").lt(50),
        palette="Purples",
        domain=[0, 50],  # not influenced by `rows`
        na_color="lightgreen",
    )
)
```
