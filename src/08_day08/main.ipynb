```python
import polars as pl
import polars.selectors as cs
from great_tables import GT, style, loc
from datetime import date, time, timedelta
```

```python
df_demo = pl.DataFrame(
    {
        "int": [1, 2, 3],
        "float1": [4.4, 5.5, 6.6],
        "float2": [7.7, 8.88, 9.999],
        "perc": [0.5401, -0.0895, 0.7946],
        "sci": [10001.001, -20002.002, 30003.003],
        "group": ["A", "B", "B"],
        "str": ["Python", "Ruby", "Rust"],
        "date": pl.date_range(date(2024, 9, 1), date(2024, 9, 3), eager=True),
        "time": pl.time_range(
            start=time(21, 0),
            interval=timedelta(hours=1, minutes=10),
            eager=True,
        ),
    }
).with_columns(
    pl.col("date").dt.combine(pl.col("time")).alias("datetime"),
)
df_demo
```

## tab_style

```python
# all columns
(
    GT(df_demo).tab_style(
        style=style.fill(color="lightgray"), locations=loc.body()
    )
)
```

```python
# using column name to select the columns
(
    GT(df_demo).tab_style(
        style=style.fill(color="lightgray"), locations=loc.body(columns="str")
    )
)
```

```python
# using cs to select columns
(
    GT(df_demo).tab_style(
        style=style.fill(color="lightgray"),
        locations=loc.body(
            columns=cs.starts_with("date") | cs.starts_with("int")
        ),
    )
)
```

```python
# using expression to select rows
(
    GT(df_demo).tab_style(
        style=style.fill(color="lightgray"),
        locations=loc.body(columns="str", rows=pl.col("group").eq("B")),
    )
)
```

```python
(
    GT(df_demo).tab_style(
        style.text(color="green", font="Georgia", weight="bold"),
        locations=loc.body(columns="str", rows=pl.col("group").eq("B")),
    )
)
```

```python
(
    GT(df_demo).tab_style(
        style.borders(color="red", style="dashed", weight="3px"),
        locations=loc.body(columns="str", rows=pl.col("group").eq("B")),
    )
)
```

```python
# all together - multiple styles with various selection methods
(
    GT(df_demo).tab_style(
        style=[
            style.fill(color="lightgray"),
            style.text(color="green", font="Georgia", weight="bold"),
            style.borders(color="red", style="dashed", weight="3px"),
        ],
        locations=loc.body(
            columns=[cs.numeric() - cs.integer(), pl.col("group"), "str", -1],
            rows=pl.col("group").eq("A").or_(pl.col("float2").gt(9)),
        ),
    )
)
```
